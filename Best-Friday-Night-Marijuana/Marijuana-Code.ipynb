```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# File import path
cannabis_file = './Resource/cannabis.csv'
```

```python
# Read dataset & remove rows with missing data
cannabis_df = pd.read_csv(cannabis_file)
cannabis_df = cannabis_df.dropna(how = 'any')
cannabis_df = cannabis_df.drop(cannabis_df[cannabis_df['Effects'] == 'None'].index)
cannabis_df = cannabis_df.drop(cannabis_df[cannabis_df['Flavor'] == 'None'].index)
cannabis_df = cannabis_df.drop(cannabis_df[cannabis_df['Description'] == 'None'].index)
cannabis_df = cannabis_df.set_index('Strain')
cannabis_df
```

```python
#List only sativa strains & rating of 3.0 or greater
sativa_strains = cannabis_df.loc[cannabis_df['Type'] == 'sativa', :]
sativa_strains = sativa_strains.sort_values('Rating', ascending = False)
best_sativa_strains = sativa_strains.loc[sativa_strains['Rating'] > 3.0]
best_sativa_strains
```

```python
#List only indica strains & rating of 3.0 or greater
indica_strains = cannabis_df.loc[cannabis_df['Type'] == 'indica', :]
indica_strains = indica_strains.sort_values('Rating', ascending = False)
best_indica_strains = indica_strains.loc[indica_strains['Rating'] > 3.0]
best_indica_strains
```

```python
#List only hybrid strains & rating of 3.0 or greater
hybrid_strains = cannabis_df.loc[cannabis_df['Type'] == 'hybrid', :]
hybrid_strains = hybrid_strains.sort_values('Rating', ascending = False)
best_hybrid_strains = hybrid_strains.loc[hybrid_strains['Rating'] > 3.0]
best_hybrid_strains
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# File import path
cannabis_file = './Resource/cannabis.csv'
```

```python
# Read dataset & remove rows with missing data
cannabis_df = pd.read_csv(cannabis_file)
cannabis_df = cannabis_df.dropna(how = 'any')
cannabis_df = cannabis_df.drop(cannabis_df[cannabis_df['Effects'] == 'None'].index)
cannabis_df = cannabis_df.drop(cannabis_df[cannabis_df['Flavor'] == 'None'].index)
cannabis_df = cannabis_df.drop(cannabis_df[cannabis_df['Description'] == 'None'].index)
cannabis_df = cannabis_df.set_index('Strain')
cannabis_df.head()
```

```python
sativa_strain = cannabis_df.loc[cannabis_df['Type'] == 'sativa', :]
sativa_strain[['Effects', 'Flavor']]
```

```python
indica_strain = cannabis_df.loc[cannabis_df['Type'] == 'indica', :]
indica_strain[['Effects', 'Flavor']]
```

```python
hybrid_strain = cannabis_df.loc[cannabis_df['Type'] == 'hybrid', :]
hybrid_strain[['Effects', 'Flavor']]
```

```python
short_cannabis_df = cannabis_df[['Type', 'Rating']]
short_cannabis_df.head()
```

```python
#List of ratings 4.5 to 5.0 by strain type
sativa_strains = short_cannabis_df.loc[short_cannabis_df['Type'] == 'sativa', :]
best_sativa_strains = sativa_strains.loc[sativa_strains['Rating'] >= 4.5]
best_sativa_count = best_sativa_strains.value_counts()

indica_strains = short_cannabis_df.loc[short_cannabis_df['Type'] == 'indica', :]
best_indica_strains = indica_strains.loc[indica_strains['Rating'] >= 4.5]
best_indica_count = best_indica_strains.value_counts()

hybrid_strains = short_cannabis_df.loc[short_cannabis_df['Type'] == 'hybrid', :]
best_hybrid_strains = hybrid_strains.loc[hybrid_strains['Rating'] >= 4.5]
best_hybrid_count = best_hybrid_strains.value_counts()
```

```python
#List of ratings 4.0 to 4.4 by strain type
better_sativa_strains = sativa_strains.loc[sativa_strains['Rating'] <= 4.4][sativa_strains['Rating'] >= 4.0]
better_sativa_count = better_sativa_strains.value_counts()

better_indica_strains = indica_strains.loc[indica_strains['Rating'] <= 4.4][indica_strains['Rating'] >= 4.0]
better_indica_count = better_indica_strains.value_counts()

better_hybrid_strains = hybrid_strains.loc[hybrid_strains['Rating'] <= 4.4][hybrid_strains['Rating'] >= 4.0]
better_hybrid_count = better_hybrid_strains.value_counts()
```

```python
#List of ratings 3.5 to 3.9 by strain type
okay_sativa_strains = sativa_strains.loc[sativa_strains['Rating'] <= 3.9][sativa_strains['Rating'] >= 3.5]
okay_sativa_count = okay_sativa_strains.value_counts()

okay_indica_strains = indica_strains.loc[indica_strains['Rating'] <= 3.9][indica_strains['Rating'] >= 3.5]
okay_indica_count = okay_indica_strains.value_counts()

okay_hybrid_strains = hybrid_strains.loc[hybrid_strains['Rating'] <= 3.9][hybrid_strains['Rating'] >= 3.5]
okay_hybrid_count = okay_hybrid_strains.value_counts()
```

```python
#List of ratings less than 3.5 by strain type
worst_sativa_strains = sativa_strains.loc[sativa_strains['Rating'] < 3.5][sativa_strains['Rating'] >= 3.0]
worst_sativa_count = worst_sativa_strains.value_counts()

worst_indica_strains = indica_strains.loc[indica_strains['Rating'] < 3.5][indica_strains['Rating'] >= 3.0]
worst_indica_count = worst_indica_strains.value_counts()

worst_hybrid_strains = hybrid_strains.loc[hybrid_strains['Rating'] < 3.5][hybrid_strains['Rating'] >= 3.0]
worst_hybrid_count = worst_hybrid_strains.value_counts()
```

```python
#Set x & y axis
x_axis = [4.5, 4.6, 4.7, 4.8, 4.9, 5.0]

#Plot on line graph
plt.plot(x_axis, best_sativa_count, color = 'blue', label = 'Sativa')
plt.plot(x_axis, best_indica_count, color = 'purple', label = 'Indica')
plt.plot(x_axis, best_hybrid_count, color = 'green', label = 'Hybrid')
plt.legend(loc = 'best')
plt.title('Best Strains per Type')
plt.xlabel('Ratings')
plt.ylabel('Number of Ratings')
plt.show()
```

```python
#Set x & y axis
x_axis = [4.0, 4.1, 4.2, 4.3, 4.4]

#Plot on line graph
plt.figure(figsize = (4,3))
plt.plot(x_axis, better_sativa_count, color = 'blue', label = 'Sativa')
plt.plot(x_axis, better_indica_count, color = 'purple', label = 'Indica')
plt.plot(x_axis, better_hybrid_count, color = 'green', label = 'Hybrid')
plt.legend(loc = 'best')
plt.title('Better Strains per Type')
plt.xlabel('Ratings')
plt.ylabel('Number of Ratings')
plt.show()
```
